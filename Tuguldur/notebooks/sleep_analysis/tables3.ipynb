```python
import pandas as pd

fname1 = "/home/nero/phasic_tonic/data/analysis_output/segmented_posttrial5/sleep_states_per_trial.csv"
fname2 = "/home/nero/phasic_tonic/data/analysis_output/whole_posttrial5/sleep_states_per_trial.csv"
```

```python
df_seg = pd.read_csv(fname1, index_col=0)
df_whole = pd.read_csv(fname2, index_col=0)
df_seg
```

```python
sleep_norm = df_seg['total_sleep_duration']/(df_seg['total_sleep_duration']+df_seg['total_awake_duration'])
df_seg.insert(13, 'sleep_normalized', sleep_norm)
df_seg
```

```python
sleep_norm = df_whole['total_sleep_duration']/(df_whole['total_sleep_duration']+df_whole['total_awake_duration'])
df_whole.insert(13, 'sleep_normalized', sleep_norm)
df_whole
```

```python
total_trial_time = (df_seg['total_sleep_duration']+df_seg['total_awake_duration'])
counts_wake_norm = df_seg['counts_wake']/total_trial_time
counts_nrem_norm = df_seg['counts_nrem']/total_trial_time
counts_rem_norm = df_seg['counts_rem']/total_trial_time
```

```python
df_seg['counts_wake_norm']=counts_wake_norm
df_seg['counts_nrem_norm']=counts_nrem_norm
df_seg['counts_rem_norm']=counts_rem_norm
df_seg
```

```python
total_trial_time = (df_whole['total_sleep_duration']+df_whole['total_awake_duration'])
counts_wake_norm = df_whole['counts_wake']/total_trial_time
counts_nrem_norm = df_whole['counts_nrem']/total_trial_time
counts_rem_norm = df_whole['counts_rem']/total_trial_time
```

```python
df_whole['counts_wake_norm']=counts_wake_norm
df_whole['counts_nrem_norm']=counts_nrem_norm
df_whole['counts_rem_norm']=counts_rem_norm
df_whole
```

```python
df_seg.to_csv(fname1, index=False)
df_whole.to_csv(fname2, index=False)
```
