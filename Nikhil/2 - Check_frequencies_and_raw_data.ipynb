# Filter + Check Spectrum/Raw Signal

```python
hpc_data = np.concatenate([hpc_data1, hpc_data2, hpc_data3, hpc_data4, hpc_data5])
pfc_data = np.concatenate([pfc_data1, pfc_data2, pfc_data3, pfc_data4, pfc_data5])
```

Select IIR or FIR filter to see their effect on the raw signal. FIR can cause cycles to appear when the signal is zero, this is an edge effect. It also takes more time than IIR filter. But IIR filter can cause phase distortion.

```python
fs = 2500

# IIR Filter - Delta band pass
hpc_data_filtered = filter_signal(hpc_data, fs, 'bandpass', (0.1, 4), n_cycles=3, filter_type='iir', butterworth_order=6, remove_edges=False)
pfc_data_filtered = filter_signal(pfc_data, fs, 'bandpass', (0.1, 4), n_cycles=3, filter_type='iir', butterworth_order=6, remove_edges=False)

# FIR Filter - Delta band pass
# hpc_data_filtered = filter_signal(hpc_data, fs, 'bandpass', (0.1, 4), n_cycles=3, filter_type='fir', remove_edges=False)
# pfc_data_filtered = filter_signal(pfc_data, fs, 'bandpass', (0.1, 4), n_cycles=3, filter_type='fir', remove_edges=False)
```

Looking at the Spectrum:

```python
n = len(pfc_data_filtered)
fft_output = np.fft.fft(pfc_data_filtered)
frequencies = np.fft.fftfreq(n, d=1/fs)
positive_frequencies_pfc = frequencies[:n // 2]
magnitude_pfc = np.abs(fft_output)[:n // 2] * 2 / n

n = len(hpc_data_filtered)
fft_output = np.fft.fft(hpc_data_filtered)
frequencies = np.fft.fftfreq(n, d=1/fs)
positive_frequencies_hpc = frequencies[:n // 2]
magnitude_hpc = np.abs(fft_output)[:n // 2] * 2 / n
```

```python
plt.subplot(2, 1, 1)
plt.plot(positive_frequencies_pfc[:fs*50], magnitude_pfc[:fs*50])
plt.title("PFC")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(positive_frequencies_hpc[:fs*50], magnitude_hpc[:fs*50])
plt.title("HPC")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid()

plt.tight_layout()
# plt.savefig(f'./plots/ORN_Frequency_Spectrum/Rat_{rat}_Spectrum')
```

Checking Raw Signal:

```python
length = 50000
start = np.random.randint(0, min(len(pfc_data), len(hpc_data)) - length)
time = [(t + start) / fs for t in range(length)]

fig = plt.figure(figsize=(10, 8))
plt.subplot(2, 1, 1)
plt.plot(time, pfc_data[start:start + length], label="Raw Data")
plt.plot(time, pfc_data_filtered[start:start + length], label="Filtered Data (0.1-4 Hz)")
plt.ylabel("Potential")
plt.xlabel("Time (sec)")
plt.title(f"Rat-{rat} PFC - {length//fs} seconds")
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(time, hpc_data[start:start + length], label="Raw Data")
plt.plot(time, hpc_data_filtered[start:start + length], label="Filtered Data (0.1-4 Hz)")
plt.ylabel("Potential")
plt.xlabel("Time (sec)")
plt.title(f"Rat-{rat} HPC - {length//fs} seconds")
plt.legend()

plt.tight_layout()
# plt.savefig(f'./plots/10-02-2025/Rat_{rat}_signal')
```

```python
# Plot raw signal with corresponding filtered signal
def plot_compare_signal(raw_signal1, raw_signal2, filt_signal1, filt_signal2, label1, label2, length=10000):
    start = np.random.randint(0, min(len(raw_signal1), len(raw_signal2)) - length)
    time = [(t + start) / fs for t in range(length)]
    
    fig = plt.figure(figsize=(10, 8))
    plt.subplot(2, 1, 1)
    plt.plot(time, raw_signal2[start:start + length], label="Raw Data")
    plt.plot(time, filt_signal2[start:start + length], label="Filtered Data (0.1-4 Hz)")
    plt.axhline(y=0.0, color='k', linestyle='--')
    plt.ylabel("Potential")
    plt.xlabel("Time (sec)")
    plt.title(label2)
    plt.legend()

    plt.subplot(2, 1, 2)
    plt.plot(time, raw_signal1[start:start + length], label="Raw Data")
    plt.plot(time, filt_signal1[start:start + length], label="Filtered Data (0.1-4 Hz)")
    plt.axhline(y=0.0, color='k', linestyle='--')
    plt.ylabel("Potential")
    plt.xlabel("Time (sec)")
    plt.title(label1)
    plt.legend()
    plt.tight_layout()
```

```python
length=10000
plot_compare_signal(pfc_data,hpc_data,
                    pfc_data_filtered,hpc_data_filtered,
                    f"Rat-3 PFC - {4} seconds",f"Rat-1 PFC - {4} seconds",length=length)
# plt.savefig(f'./plots/ORN_Frequency_Spectrum/signal.png')
```
